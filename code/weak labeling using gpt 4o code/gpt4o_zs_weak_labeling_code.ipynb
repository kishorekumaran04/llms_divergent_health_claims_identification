```python
import time
import openai
```

```python
api_key = ''
with open('chat_gpt_api_key.txt') as f:
    lines = f.read()
    api_key = lines
```

```python
print(api_key)
```

```python
from openai import OpenAI
client = OpenAI(api_key=api_key)
```

```python
def chat(inp, role='user', temp=0.0, top_p=1.0):
    '''
    Call ChatGPT API
    '''
    message = [{'role': role, 'content': inp}]
    while True:
        try:
            completion = client.chat.completions.create(
                model='gpt-4o',
                # model='gpt-3.5-turbo',
                messages=message,
                temperature=temp,
                top_p=top_p
            )
            reply_content = completion.choices[0].message.content
            return reply_content
        except:
        #   print("Exception")
          time.sleep(2)
          continue
```

```python
print(chat("tell me a story containing atmost 100 characters"))
```

```python
from datetime import datetime
import pandas as pd
```

```python
def prompt_zero_shot(sentence1, sentence2):
    prompt = "Classify if the relationship between two statements is divergent or not:\nStatement 1: {}\nStatement 2: {}\nYour final answer should only contain the word 'Divergent' or 'Not Divergent'".format(sentence1, sentence2)
    return prompt
```

```python
df = pd.read_csv('./combined_12.csv')
df.head()
```

```python
prompt_zero_shot(df.iloc[0]["text1"], df.iloc[0]["text2"])
```

```python
len(df)
```

```python
# Process test with gpt-4o zs prompt
for i in range(len(df)):
    print(i)
    row = df.iloc[i]
    sentence1 = row["text1"]
    sentence2 = row["text2"]
    gpt_4o_zs = chat(prompt_zero_shot(sentence1, sentence2))

    df.loc[i] = [sentence1, sentence2, gpt_4o_zs]

print("Done")



```

```python
final_value = [""]*len(df)
for i in range(len(df)):
  temp = df.iloc[i]['ZS response']
  if "Not Divergent" in temp:
    final_value[i] = "Not Divergent"
  elif "Divergent" in temp:
    final_value[i] = "Divergent"
```
