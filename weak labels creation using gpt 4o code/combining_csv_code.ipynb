```python
# code to remove duplicates

def check_equal(row1, row2):
    if (row1["text1"] == row2["text1"]) & (row1["text2"] == row2["text2"]):
        return True
    elif (row1["text1"] == row2["text2"]) & (row1["text2"] == row2["text1"]):
        return True
    else:
        return False


def clean_dataframe(df, start_index):
    print(start_index)
    call_recursion = False
    # for i in range(len(df)):
    i = start_index
    new_df = df.copy()
    while i<len(df):
        row1 = df.iloc[i]
        # print(row1["text1"])
        indices = df.index[((row1["text1"] == df["text1"]) | (row1["text1"] == df["text2"])) & (df.index > i)] 
        # print(indices)
        for index in indices:
            # print(index)
            row2 = df.iloc[index]
            if check_equal(row1, row2):
                # print(str(i) + "  " + str(j))
                new_df = new_df.drop(index)
                # print(new_df)
                call_recursion = True
        i+=1
        if call_recursion:
            break
    if call_recursion:
        new_df.reset_index(drop=True, inplace=True)
        return clean_dataframe(new_df, start_index = i)
    else:
        return df

```

```python
df_1 = pd.read_csv('./combined_all with gpt4o zs response.csv')
new_df_1 = clean_dataframe(df_1, start_index=0)
csv_data_1 = new_df_1.to_csv('./combined_all with gpt4o zs response no duplicates.csv', index = False)
```
